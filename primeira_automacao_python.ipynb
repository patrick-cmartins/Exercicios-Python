```python
# Passo 1: Entrar no sistema (link do Google Drive)
import pyautogui as gui
import time
import pyperclip as clip

gui.PAUSE = 1

gui.alert("O programa via começar. Não use nada no seu computador enquanto isso")
# Passo 1: Entrar no sistema (link do Google Drive)

#gui.press("win")
#gui.write("brave")
#gui.press("enter")
#time.sleep(5)
#gui.write("drive")
#gui.press("tab")
#gui.press("enter")
gui.hotkey("ctrl","t")
link = "https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym"
clip.copy(link)
gui.hotkey("ctrl", "v")
gui.press("enter")
time.sleep(5)

# Passo 3: Fazer o download da Base de Vendas

# time.sleep(5) // time.sleep junto com o position serve para vc saber qual a posição do mouse que vc quer clickar
# gui.position()

gui.click(478, 434, clicks = 1)
gui.click(1705, 185, clicks = 1)
gui.click(1557, 561, clicks = 1)
time.sleep(5)

# Passo 4: Calcular os indicadores (Faturamento e a Quantidade de Produtos)
# formatar a leitura do pandas com "r" que significa rockstring ( leia esse texto exatamente como esta escrito )
# no python tem muito código secreto para realizar outras tarefas
import pandas as pd
tabela = pd.read_excel(r"D:\Download\Vendas - Dez.xlsx")
# tabela = pd.read_excel(r"D:\Download\Vendas - Dez.xlsx", sheet_name = "Qual aba que vc deseja selecionar dentro do excel")
display(tabela)
faturamento = tabela["Valor Final"].sum()
qtde_produtos = tabela["Quantidade"].sum()

```

```python
# Passo 5: Entrar no meu email
gui.hotkey("ctrl", "t")
linkmail = "gmail.com"
clip.copy(linkmail)
gui.hotkey("ctrl", "v")
gui.press("enter")
time.sleep(10)
# Passo 6: Criar o email
gui.click(108, 221, clicks = 1)
maildiretoria = "patrick.cmartins01@gmail.com"
clip.copy(maildiretoria)
gui.hotkey("ctrl", "v")
time.sleep(2)
gui.press("tab")
relatorio = "Relatório de vendas"
# quando for digitar algo com caractere especial é necessario declarar a frase como variavel, pois o python nao digita numeros com caracteres
clip.copy(relatorio)
gui.hotkey("ctrl","v")
gui.press("tab")
texto_email = f"""
Prezados, bom dia

O faturamento foi de R$:{faturamento:,.2f}
A quantidade de produtos foi de:{qtde_produtos:,.2f}

Abs
Patrick Martins"""
clip.copy(texto_email)
gui.hotkey("ctrl","v")
# biblioteca OS para saber quais os ultimos itens de uma pasta para saber qual foi o ultimo download
# Passo 7: Enviar o email
gui.hotkey("ctrl", "enter")

gui.alert("Processo completo!!")
```
